# How I'm Addressing These Gaps
Key takeaways from videos

## 3.11 Binary Search – Daily Video 1
- Understands how binary search splits a sorted list in half repeatedly
- Compares the middle item to the target
- Efficient for large datasets (logarithmic time)
- Only works on sorted lists

![](/alex_2025/images/notebooks/foundation/binsearch.png)

## 3.17 Algorithmic Efficiency - Daily Video 1 & 2
- Measures efficiency using time complexity (Big O)
- Reasonable vs. unreasonable runtimes (e.g. linear vs. exponential)
- Impacts scalability of programs
- Heuristics: useful when exact solutions are impractical
- Sacrifices precision for speed
- Common in optimization and AI problems

![](/alex_2025/images/notebooks/foundation/reason.png)

## 3.18 Undecidable Problems – Daily Video 1
- Some problems have no algorithmic solution
- The Halting Problem: famous example
- Recognize the limits of computation
- Not all problems are solvable by code

![](/alex_2025/images/notebooks/foundation/halting.png)

## 5.2 Digital Divide – Daily Video 1
- Gap in access to technology and the internet
- Affects education, job opportunities, civic participation
- Often linked to income, geography, age

![](/alex_2025/images/notebooks/foundation/divide.png)

## 5.3 Computing Bias – Daily Video 1
- Algorithms can reinforce human biases
- Training data matters
- Bias in AI can lead to unfair outcomes
- Need for fairness and transparency in computing

![](/alex_2025/images/notebooks/foundation/bias.png)
