## Project Feature Writeup

<br>

The customizable profile page allows users to **input** and manage personal information, such as name, username, email, password, date of birth, and city, through editable input fields. Users interact with these fields by clicking to edit the data, and upon submitting the changes, the updated information is **stored in a database** using Flask and SQLAlchemy. A database is used to manage the collection of user data, and the program includes a **procedure**, put(), which accepts **parameters** (user ID, name, username, email, password, dob, city), validates the input, updates the database with the new information, **and returns a success or error message** based on the validity of the input. The procedure incorporates **sequencing** (processing inputs), **selection** (checking for missing or invalid data), and **iteration** (iterating over input fields for validation). After the user submits their changes, the updated profile is displayed on the frontend, with **visual output** showing the updated data, and a confirmation message is provided to inform the user that their profile has been successfully updated.
